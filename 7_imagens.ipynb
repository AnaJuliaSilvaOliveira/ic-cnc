```python
import matplotlib.pyplot as plt
import pandas as pd
plt.rcParams.update({'font.size': 13})

original_ = pd.read_csv('split_cutting_tests_processed/2inch_stickout/c_320_005_split_3.csv')
noise = pd.read_csv('split_cutting_tests_processed/2inch_stickout/c_320_005_split_3_noise_added.csv')

plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
plt.xlabel('Dislocation')
plt.ylabel('Time')
plt.title('Original file without noise')
plt.plot(original_['t'],original_['y'],c='dimgray')

plt.subplot(1,2,2)
plt.xlabel('Dislocation')
plt.ylabel('Time')
plt.title('Modified file with Gaussian noise')
plt.plot(noise['t'],noise['y'],c='dimgray')

plt.tight_layout()
plt.show()
```
